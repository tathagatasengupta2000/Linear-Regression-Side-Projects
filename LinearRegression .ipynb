```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df=pd.read_csv('population deficit.csv')
df
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Death")
plt.scatter(df.Year,df['Population Deficit'],color='Blue',marker="+")

```

```python
reg=LinearRegression()
reg.fit(df[['Year']],df[['Population Deficit']])
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d=pd.read_csv('Year.csv')
d
```

```python
prediction=reg.predict(d)
```

```python
d['Population Deficit']=prediction
d
```

```python
d.to_csv("Complete Prediction.csv")
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Death")
plt.scatter(df.Year,df['Population Deficit'],color='Blue',marker="+")
plt.plot(df.Year,reg.predict(df[['Year']]),color="Red")
```
